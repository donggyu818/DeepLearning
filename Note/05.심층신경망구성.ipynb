#### 패션 MNIST

```python
from tensorflow import keras

(train_input, train_target), (test_input, test_target) = \
    keras.datasets.fashion_mnist.load_data()
```

```python
# train 과 valid data 분리
from sklearn.model_selection import train_test_split

train_scaled = (train_input / 255.0).reshape(-1, 28*28)     # 정규화 + 2차

train_scaled, val_scaled, train_target, val_target = \
    train_test_split(
        train_scaled,
        train_target,
        test_size=0.2,
        random_state=42
    )
```

#### Layer 를 추가하는 방법 - 1

```python
# Layer 만들기
dense1 = keras.layers.Dense(100, activation='sigmoid', input_shape=(784,))  # 은닉층
dense2 = keras.layers.Dense(10, activation='softmax')   # 출력층. 다중 출력이면 거의 softmax
```

```python
# 전체 모델에 층을 추가
model = keras.Sequential([dense1, dense2])
```

```python
# Summary
model.summary()
```

----
#### Layer 를 추가하는 방법 - 2

```python
model = keras.Sequential(
    [
        keras.layers.Dense(100, activation='sigmoid', input_shape = (784,), name = 'hidden'),
        keras.layers.Dense(10, activation='softmax', name='output')
    ],
    name='패션 MNIST 모델'
)
```

```python
model.summary()
```

----
#### Layer 를 추가하는 방법 - 3

```python
model = keras.Sequential()
model.add(keras.layers.Dense(100, activation='sigmoid', input_shape = (784, ), name='hidden'))
model.add(keras.layers.Dense(10, activation='softmax', name='output'))
```

```python
model.summary()
```

> 제일 많이 사용하는 방법. 파이썬 코드가 중간에 삽입이 가능하기 때문. (ex_if문으로 분기 줄 수 있음)

```python
model.compile(
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
model.fit(train_scaled, train_target, epochs=5)
```

-----
#### 입력층을 이미지 그대로 사용하기
(train_scaled = (train_input / 255.0).reshape(-1, 28*28) 에서 reshape 를 안 거치고 만들기)

```python
(train_input, train_target), (test_input, test_target) = \
    keras.datasets.fashion_mnist.load_data()

train_scaled = train_input / 255.0

train_scaled, val_scaled, train_target, val_target = \
    train_test_split(
        train_scaled,
        train_target,
        test_size=0.2,
        random_state=42
    )
```

```python
train_input.shape
```

```python
model = keras.Sequential()

model.add(keras.layers.Flatten(input_shape = (28, 28)))   # flatten : 펴주는 함수
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))     # 둘 중 하나면 시그모이드
```

```python
model.summary()
```

```python
# 손실함수
import numpy as np

model.compile(
    optimizer = 'adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# 학습하기
model.fit(np.array(train_scaled), np.array(train_target), epochs = 16)     # 현재 pandas 데이터인데 numpy 로 바꿔서 훈련해야 함
```

```python
# 검증 데이터로 확인
print(model.evaluate(val_scaled, val_target)[1])
```

```python
# 테스트로 확인 . 테스트는 / 255 를 안 함. 정규화 해줘야 함.
test_scaled = test_input / 255.0
print(model.evaluate(train_scaled, train_target)[1])
print(model.evaluate(test_scaled, test_target)[1])
```
