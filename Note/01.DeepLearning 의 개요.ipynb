### Deep Learning
- 머신러닝의 한 종류
- 인공신경망은 인간의 뇌신경망의 신호 체계를 본따서 만든 구조
- 이미지 인식, 음성 인식, 자연어 처리 등의 다양한 분야에 활용

#### Deep Learning 과 Machine Learning 의 **차이점**
- 가장 큰 차이점은 '특징량(Feature) 추출' 이다.
- 머신러닝은 사람(분석가)이 이미지 내부에 있는 특징을 지정하는 방식
- 딥러닝에서는 학습데이터에서 기계가 자동으로 특징을 추출
- Deep Learning 은 스스로 학습을 하기 때문에 Machine Learning 보다 많은 데이터가 필요함
- 즉, 머신러닝까지는 인간이 하나하나 가르쳐야 했지만 딥러닝을 이용하면 기계가 자동으로 특징을 학습할 수 있다.

#### 신경망(Neural Network)
- 인간의 신경망을 본따 만든 네트워크 구조를 의미
- 입력층(데이터 불러옴, Feature), 은닉층(숨겨진 애), 출력층(데이터 출력)을 지나 처리가 일어나고 최종적인 결과가 출력된다.
- 이러한 신경망을 은닉층의 2개 이상 추가되면 깊은 신경망(Deep Neural Network)이라고 하며 이를 활용한 기계학습을 '딥러닝'이라고 부른다.
- 딥러닝은 대량의 데이터를 입력해서 학습한다.
- 학습이란 유닛(Unit)의 접속 가중치를 조정하는 것을 의미. (접속 가중치 : y = ax + b 중에 a / b 가 가중치임)

#### Deep Learning 이 Machine Learning 보다 우수한 분야
- Image Classification
- Image Object detection
> 발전된 분야 : 자율 주행차
- Sequence data 분석 및 예측(순서가 있는 데이터)
> 발전된 분야 : 번역기(어순이 있음)

### 용어 정의

#### 손실함수(Loss Function)
- 출력값(예측값)과 정답의 차이 ( |y - y^|)
- 출력값과 정답이 일치 할수록 손실함수의 값은 작아진다.
- 회귀에는 평균절대오차(MAE)와 평균제곱오차(Mean Squared Error: MSE)
- 분류 문제에는 크로스 엔트로피(Cross Entropy)를 사용한다.
- 매개변수(Weight, Bias) 를 조절해서 손실함수의 값을 최저로 만드는 과정을 최적화(Optimization)이라고 한다.
- 최적화 과정은 Optimizer 를 통해 이루어지며 Optimizer 는 역전파(back propagation) 과정을 수행해서 딥러닝 모델의 매개변수를 최적화 한다.

#### 최적화(Optimization)
- 대표적인 방법은 경사하강법(Batch Gradient Descent)
- 반복적으로 손실함수에 대한 모델 매개변수의 미분값(기울기)을 구한 후 그 미분값의 반대방향으로 매개변수를 조절해 나가면 결국에는 최저 손실함수에 도달한다.
